First, we want to import matplotlib, a common python plotting package (https://matplotlib.org/). 'plt' is a common abbreviation used for matplotlib. When we import packages, we often want to import them by shorter names in order for faster/easier coding.

astropy (https://www.astropy.org/) is the most commonly used Python package for dealing with astrophysical data.

```python
#importing matplotlib by its shorter, common abbreviation 'plt'
import matplotlib.pyplot as plt
```

The next two lines of code are not necessary! You can experiment with or without them

```python
#from astropy's visualization tools, we are importing the common matplotlib
#style 
from astropy.visualization import astropy_mpl_style

#style.use is a matplotlib method to use your own matplotlib style instead of the
#standard settings
plt.style.use(astropy_mpl_style)
```

We are using a photo of the horsehead nebula for variety (https://en.wikipedia.org/wiki/Horsehead_Nebula). Supposedly it resembles a horse's head, but... idk about that one. Still very cool though.

```python
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits

image_file = get_pkg_data_filename('tutorials/FITS-images/HorseHead.fits')
```

```python
#here, ext = 0 is refering to the primary HDU (which stands for header data units)
#we want to extract the image data from the fits file
image_data = fits.getdata(image_file, ext=0)
```

These are standard matplotlib plotting modes, plt.figure calls for a figure object. If you want to mess around with your image further, go to https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html for more figure parameters. 

For example, if you want to increase/decrease the figure size or make it a rectangle instead of a square, you would do plt.figure(figsize = (6,6)) and change the numbers.

If you want to change the figure size and change the background color, you would do plt.figure(figsize = (6,7),facecolor='red'), separating the two with a comma. For a list of matplotlib colors, go to this page, past the block of code https://matplotlib.org/3.1.0/gallery/color/named_colors.html.

For a list of colormaps, scroll to the bottom of this page https://matplotlib.org/tutorials/colors/colormaps.html and there is a list of colormap names along with what they look like. Experiment with what looks best to you!

The documentation of text in matplotlib can be found here https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.text.html for additional parameters aside from the ones I've already shown. One of those is adding a box around your text, along with others. Feel free to mess around with this.

Matplotlib has many other plotting features that are useful, such as adding lines, arrows, shapes to the figure. Feel free to mess around with it!

```python
plt.figure()

#this line of code plots the image and shows the colormap
plt.imshow(image_data, cmap='magma')

#matplotlib has many functionalities, one of those is text. The first two arguments of plt.text are x and y coordinates
#the next argument is a string of what you want the text to say. The other two are additional arguments of color (c)
#and size.
plt.text(100,100,'hi',c='white',size=20)

#this line is showing the colorbar, the units used by the color bar are units of flux or brightness
#the x and y axes are referencing the pixel positions
plt.colorbar()

#to save your figure, you don't need to take a screenshot. You can just use the following line of code to save the image
#the argument it takes is a string of what you want to call the file. Once you run this block of code, it will save the
#image to your desktop wherever you are running this python file. That's why I've commented the line out, so you can
#experiment with different settings. Once you're done, un-comment the line below (remove the hashtag symbol)

#plt.savefig('horseheadnebula.png')
```
